# **Instance Variable & Instance Method**
- **Instance**: 
- refers to an object created from a class.
- each object has its own attributes and methods

```python
class Student:
    def __init__ (self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade
    
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Grade: {self.grade}")
    
    def is_eligible(self):
        if self.age >= 15:
            print(f"{self.name} is eligible for the exam.")
        else:
            print(f"{self.name} is not eligible for the exam.")              
```

```python
s1 = Student("Ali", 16, "A")
s2 = Student("Ahmed", 13, "B")

s1.display_info()
s1.is_eligible()

s2.display_info()
s2.is_eligible()
```
